# 2.1.1 Ejemplo de puerta AND
---

Bien, empecemos viendo un ejemplo de como crear una red muuuuy sencilla para modelar el comportamiento de una **puerta lógica AND**. En este ejemplo utilizaremos un **perceptrón**, es decir, una única neurona, como vimos en clase. Para ello, hay que seguir los siguientes pasos:

```python
# Lo primero de todo es importar nuestra libreria 
# de Tensorflow, para ello escribimos
import tensorflow as tf

# Tambien importaremos la libreria de Numpy
import numpy as np


# 1)
# Lo siguiente seria preparar el dataset que queremos entrenar, es decir, los 
# datos que usaremos como ejemplo para que la red aprenda.

# Como lo que queremos modelar es el comportamiento de una puerta logica AND,
# el dataset en este caso es muy reducido y sencillo. Tan solo hay que coger 
# todas las posibles combinaciones que tien una puerta logica AND y 
# meterlas dentro de dos listas, una con los datos de entrada (datos de ejemplo)
# y otra con los de salida (soluciones).

# A la primera lista con los datos de ejemplo la llamaremos x_train
x_train = [
                [0,0],
                [0,1],
                [1,0],
                [1,1]
]

# A la segunda lista, con las soluciones a los datos la llamaremos y_train
y_train = [
                [0],
                [0],
                [0],
                [1]
]

# como ves, las entradas de las listas estan ordenadas para que cuadren los 
# datos y las soluciones
print("Soluciones de una puerta logica AND:")
for i, d in enumerate(x_train):
    print("Para la entrada ", d, " la puerta da una salida de ", y_train[i])

# Transformaremos estas listas a listas de Numpy
x_train = np.array(x_train)
y_train = np.array(y_train)



# 2)
# Una vez el dataset está preparado, podemos montar nuestra red.
# Para ello utilizaremos la funcion tf.keras.models.Sequential() la cual
# nos construira nuestra red con la estructura que le pasemos. Como nuestra 
# red la va a componer una unica neurona, solo hará falta que le pasemos una 
# cosa.

# ¿Que cosa? Pues dentro de tf.keras.layers encontraremos los distintos tipos 
# de capas que le podemos añadir a una red. De momento nosotros 
# usaremos tf.keras.layers.Dense(), funcion la cual nos prepara una capa de 
# neuronas simple.
# Solo tenemos que pasarle el numero de neuronas que queremos, 1 en este caso, y
# la funcion de activacion de esta neurona, la cual de momento sera lineal, para
# seguir los pasos vistos en clase.

# Tambien estará la capa tf.keras.Input() que solo sirve para preparar los datos
# antes de entrar a la red

model = tf.keras.models.Sequential([
  tf.keras.Input(shape=x_train[0].shape),
  tf.keras.layers.Dense(1, activation='linear')
])





# 3)
# Lo siguiente que tenemos que hacer es configurar nuestra funcion de loss. 
# ¿Para que sirve la funcion de loss? Es la manera que tenemos de transformar 
# la salida de la red en un error legible, es decir, un numero que nos diga 
# cuanto se ha equivocado la red, y asi poder utilizarlo para entrenar la red.

# Para este problema utilizaremos el error MeanSquaredError, que lo que hace es 
# calcular la diferencia entre lo que la red predice y las soluciones de los 
# datos y eleva ese numero al cuadrado, para que el signo del error no sea 
# relevante y para que errores mas grandes inflyan mucho mas en en entrenamiento.
loss_fn = tf.keras.losses.MeanSquaredError()

# Tambien tendremos que preparar el optimizador, el cual preparará la estrategia
# para modificar la red una vez sabemos cuanto se equivoca, y asi ir 
# entrenandola poco a poco.
# Aqui podemos especificarle el learning_rate del entrenamiento
opt = tf.keras.optimizers.Adam(learning_rate=0.1)




# 4)
# Antes de comenzar a entrenar nuestra red, hay que compilarla, es decir, 
# compilar el modelo que hemos creado antes. Para ello, usamos la funcion 
# compile sobre el modelo, pasandole como parametros el optimizador, la funcion 
# de loss y las metricas que se quieren medir durante el entrenamiento
model.compile(optimizer=opt,
              loss=loss_fn,
              metrics=['accuracy'])



# 5)
# ¡¡INICIAMOS EL ENTRENAMIENTO!! Para ello usamos la funcion fit() sobre el 
# modelo de la siguiente forma, pasandole el dataset de entrenamiento y el 
# numero de epocas a entrenar
model.fit(x_train, y_train, epochs=20)


"""
# Ahora nuestra red esta entrenada y sabemos que con los datos de entrenamiento 
# tiene una buena precision, pero ¿funciona tan bien realmente? Para 
# comprobarlo, vamos a utilizar los datos de validacion o test a ver que 
# precision alcanza la red con estos datos
model.evaluate(x_train,  y_train, verbose=2)
"""

# 6)
# Por ultimo, ahora que nuestra red esta entrenada, vamos a probar a pasarle 
# datos a ver que es lo que nos saca. Para ello usaremos el modelo (la variable)
# como si fuera una funcion, pasandole por parametros una lista con los datos 
# que queremos que prediga

# Los datos sera nuestra lista de datos de entrada x_train
predicciones = model(x_train).numpy()

# Mostramos la salida de la red para cada uno de los datos
for i, p in enumerate(predicciones):
    print("Para la entrada ", x_train[i], " la red predice una salida de ", p)
```

Como podemos ver, a medida que el **error** (*loss*) baja, la red hace mejor su trabajo, hasta el punto en el que la **precisión** (*accuracy*) alcanza 1.0, es decir, el 100%.

De esta manera, acabamos de entrenar nuestra primera red para que modele el comportamiento de una puerta lógica AND.


Cabe destacar que la salida hecha al final del código no es discreta, pero nosotros podríamos banalizarla manualmente, cambiándola a 1 si es superior a 0.5 y a 0 si es inferior.

# 2.1.2 Ejemplo de puerta OR
---

Y, ahora que sabemos cómo modelar una puerta lógica AND, ¿Que haría falta para modelar una **puerta lógica OR**? Veámoslo.

Las puertas lógicas AND y OR son muy similares, solo cambia entre ellas el "cuando" la puerta saca un 1 en su salida o cuando saca un 0. Con esto en mente, lo primero que se nos puede ocurrir para cambiar el comportamiento de nuestra red y cambiar los datos con los que esta se entrena. Como ahora es una puerta lógica OR, la red quedaría así:

```python
import tensorflow as tf
import numpy as np



# Montamos el dataset,  ahora de una puerta OR
x_train = [
                [0,0],
                [0,1],
                [1,0],
                [1,1]
]


y_train = [
                [0],
                [1],
                [1],
                [1]
]



# Mostramos el dataset

print("Soluciones de una puerta logica OR:")
for i, d in enumerate(x_train):
    print("Para la entrada ", d, " la puerta da una salida de ", y_train[i])

# Transformaremos estas listas a listas de Numpy
x_train = np.array(x_train)
y_train = np.array(y_train)



# Creamos la red
model = tf.keras.models.Sequential([
  tf.keras.Input(shape=x_train[0].shape),
  tf.keras.layers.Dense(1, activation='linear')
])




# Preparamos la funcion de loss y el optimizador
loss_fn = tf.keras.losses.MeanSquaredError()

opt = tf.keras.optimizers.Adam(learning_rate=0.1)



# Compilamos el modelo
model.compile(optimizer=opt,
              loss=loss_fn,
              metrics=['accuracy'])


# Entrenamos la red
model.fit(x_train, y_train, epochs=20)


# Sacamos las predicciones
predicciones = model(x_train).numpy()

# Mostramos la salida de la red para cada uno de los datos
for i, p in enumerate(predicciones):
    print("Para la entrada ", x_train[i], " la red predice una salida de ", p)
```

# 2.1.3 Ejemplo de dataset de Tensorflow
---

Como extra, veremos que dentro de la librería de `Tensorflow` hay guardados algunos **datasets** para experimentar de manera rápida y fácil con la construcción y entrenamiento de redes.

Estos **datasets** se encuentran en `tf.keras.datasets`. 

¡Veamos cómo podemos utilizarlos!


```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg



# Para acceder y cargar los datasets de tensorflow, tenemos que acceder a 
# tf.keras.datasets y, dentro, al dataset que queramos utilizar.
# En este ejemplo utilizaremos el dataset mnist

# Obtenemos el dataset como variable
mnist = tf.keras.datasets.mnist

# Preparamos el dataset, dividiendolo en las parejas de listas de 
# entrenamiento y validacion
(x_train, y_train), (x_test, y_test) = mnist.load_data()
# Aqui solo esta normalizando los datos de entrada
x_train, x_test = x_train / 255.0, x_test / 255.0

# Vemos como han quedado los datos de las listas

print("x_train: ", x_train.shape)
print("y_train: ", y_train.shape)
print("x_test: ", x_test.shape)
print("y_test: ", y_test.shape)



# El resto es igual a como hemos visto antes, montamos la red y la entrenamos
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10)
])



# Configuramos la funcion de loss.
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)



# Compilamos el modelo
model.compile(optimizer='adam',
              loss=loss_fn,
              metrics=['accuracy'])



# Entrenamos
model.fit(x_train, y_train, epochs=5)



# Evaluamos la red
model.evaluate(x_test,  y_test, verbose=2)


# Sacamos la prediccion de la red para algunas entradas
predicciones = model(x_test[:5])

fig, axs = plt.subplots(5, figsize=(10,10))
# Mostramos algunos ejemplos de la red
for i, p in enumerate(predicciones):
    # Sacamos un dato del dataset de validacion
    img = x_test[i]

    # Mostramos ese dato con matplotlib
    axs[i].imshow(img)

    # Imprimimos la prediccion
    print("La red predice que lo que se ve en la imagen", i, "es: ", np.argmax(p))

```

Como vemos, entrenar una red utilizando datasets de Tensorflow no es especialmente complicado, dependiendo del dataset.

Al final, lo que hay que hacer es preparar el dataset para ser usado y construir una red que acepte esos datos para ser entrenada.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2


# Cargamos el dataset, el cual se encuentra codificado en una imagen
dataset = cv2.imread("nube de puntos.png")
dataset = cv2.cvtColor(dataset, cv2.COLOR_BGR2RGB)

fig, axs = plt.subplots(3, figsize=(10,10))
axs[0].imshow(dataset)


# Transformamos de la imagen a las listas de entrenamiento y validacion
x_train = []
y_train = []
x_test  = []
y_test  = []

for i in range(50):
    for j in range(50):
        pixel = dataset[i][j]
        coords = [i/50,j/50]
        if pixel[2] == 0:
            # train
            if pixel[0] == 255:
                # rojo
                x_train.append(coords)
                y_train.append([1,0])
            elif pixel[1] == 255:
                # verde
                x_train.append(coords)
                y_train.append([0,1])

        else:
            # test
            if pixel[0] == 255:
                # morado
                x_test.append(coords)
                y_test.append([1,0])

            elif pixel[1] == 255:
                # cian
                x_test.append(coords)
                y_test.append([0,1])



# Re montamos la imagen a partir del dataset a ver que se ve
build_img = np.zeros(dataset.shape, dtype=dataset.dtype)

for i, x in enumerate(x_train):
    y = y_train[i]
    coordX, coordY = x
    coordX = int(round(coordX * 50))
    coordY = int(round(coordY * 50))

    if y[0] == 1:
        # Equipo rojo
        build_img[coordX][coordY] = [255, 0, 0]
    elif y[1] == 1:
        # Equipo verde
        build_img[coordX][coordY] = [0, 255, 0]
            

axs[1].imshow(build_img)



# Mostramos el dataset

print("Ejemplos del dataset:")
print("Para la entrada ", x_train[35], " el equipo es ", y_train[35])
print("Para la entrada ", x_train[36], " el equipo es ", y_train[36])
print("\n")


# Transformaremos estas listas a listas de Numpy
x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
y_test = np.array(y_test)

########################################################################
"""                       Aqui empiezas tu                           """
########################################################################







# Creamos la red
model = tf.keras.models.Sequential([
  tf.keras.Input(shape=x_train[0].shape),
  tf.keras.layers.Dense(100, activation='sigmoid'),
  tf.keras.layers.Dense(2, activation='sigmoid')
])




# Preparamos la funcion de loss y el optimizador
loss_fn = tf.keras.losses.CategoricalCrossentropy()
opt = tf.keras.optimizers.Adam(learning_rate=0.1)




# Compilamos el modelo
model.compile(optimizer=opt,
              loss=loss_fn,
              metrics=['accuracy'])


# Entrenamos la red
model.fit(x_train, y_train, epochs=150)





# Sacamos las predicciones
predicciones = model(x_test).numpy()

# Mostramos la salida de la red para cada uno de los datos
print("\n\n =======  Salidas de la red para el dataset de validacion  =======")
for i, p in enumerate(predicciones):
    print("Para la entrada ", x_test[i], " la red predice una salida de ", p, " {true: ", y_test[i], "}")


# Montamos una imagen para todas las salidas, a ver que ocurre
preds_img = np.zeros(dataset.shape, dtype=dataset.dtype)

for i in range(50):
    for j in range(50):
        coords = np.array([[i/50, j/50]])

        pred = model(coords).numpy()

        idx = np.argmax(pred)
        pixel = [0,0,0]
        pixel[idx] = 255
        preds_img[i][j] = pixel

axs[2].imshow(preds_img)

```

### Construye, entrena y muestra los resultados de un modelo que resuelva un problema de clasificación a tu elección. Comenta los pasos que vas haciendo y el porqué.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2


# Cargamos el dataset, el cual se encuentra codificado en una imagen
dataset = cv2.imread("5_equipos.png")
dataset = cv2.cvtColor(dataset, cv2.COLOR_BGR2RGB)

fig, axs = plt.subplots(2, figsize=(10,10))
axs[0].imshow(dataset)




# Equipos (valores de RGB)
team_blue   = (  0,    0,  255)
team_red    = (255,    0,    0)
team_green  = (  0,  255,    0)
team_cyan   = (  0,  255,  255)
team_purple = (255,  0,    255)

# Metemos los equipos en una lista, de forma que los equipos quedan nunmerados 
# por el orden que ocumpan en la lista
teams = [team_blue, team_red, team_green, team_cyan, team_purple]

def getTeamIndex(pixel):
    """
    Devuelve el numero identificador del equipo al que pertenece el pixel dado
    """
    global teams
    for i, team in enumerate(teams):
        if team[0] == pixel[0] and team[1] == pixel[1] and team[2] == pixel[2]:
            return i

# Transformamos de la imagen a las listas de entrenamiento y validacion
x_train = []
y_train = []
team_count = [0,0,0,0,0]
black_count = 0

for i in range(50):
    for j in range(50):
        pixel = dataset[i][j]
        coords = [i/50,j/50]
        
        team_number = getTeamIndex(pixel)


                
        if team_number != None:
            team_count[team_number] += 1

            label = [0,0,0,0,0]
            label[team_number] = 1

            x_train.append(coords)
            y_train.append(label)
        else:
            black_count += 1

print("team_count:", team_count)
print("black_count:", black_count)


########################################################################


# Mostramos el dataset

print("Ejemplos del dataset:")
print("Para la entrada ", x_train[35], " el equipo es ", y_train[35])
print("Para la entrada ", x_train[36], " el equipo es ", y_train[36])
print("\n")


# Transformaremos estas listas a listas de Numpy
x_train = np.array(x_train)
y_train = np.array(y_train)


# Creamos la red
model = tf.keras.models.Sequential([
  tf.keras.Input(shape=x_train[0].shape),
  tf.keras.layers.Dense(10, activation='sigmoid'),
  tf.keras.layers.Dense(10, activation='sigmoid'),
  tf.keras.layers.Dense(5, activation='sigmoid')
])




# Preparamos la funcion de loss y el optimizador
loss_fn = tf.keras.losses.CategoricalCrossentropy()

opt = tf.keras.optimizers.Adam(learning_rate=0.01)



# Compilamos el modelo
model.compile(optimizer=opt,
              loss=loss_fn,
              metrics=['accuracy'])


# Entrenamos la red
model.fit(x_train, y_train, batch_size=1, epochs=20)



# Montamos una imagen para todas las salidas, a ver que ocurre
preds_img = np.zeros(dataset.shape, dtype=dataset.dtype)

for i in range(50):
    for j in range(50):
        coords = np.array([[i/50, j/50]])

        pred = model(coords).numpy()

        idx = np.argmax(pred)
        pixel = teams[idx]
        preds_img[i][j] = pixel

axs[1].imshow(preds_img)

```

# 3.1) Ejemplo de dataset Tensorflow
---
---

Bien, para ver que es lo nuevo que tienen estas redes neuronales convolucionales, que mejor que verlo con un **ejemplo**.

Si recuerdas, las redes neuronales convolucionales son redes que utilizan imágenes como entrada, aplicando una operación llamada convolución para extraer las características de la imagen, es decir, lo que se ve en ella.

Pues bien, para construir una red neuronal convolucional **clasificadora**, tan solo tenemos que construir una red como lo hemos hecho hasta ahora, pero añadiendo unas capas al principio llamadas capas convolucionales (y alguna otra cosa más :p). El resto del proceso es igual.

Veamos el ejemplo:

```python

import tensorflow as tf

from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np




# Cargamos un dataset de clasificacion de imagenes. Es igual que cualquier otro 
# dataset, lo unico que cambia es que ahora la entrada son imagenes enteras
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalizamos los valores de los pixeles de las imagenes del dataset, para 
# que pasen del rango 0-255 a 0.0-1.0
train_images, test_images = train_images / 255.0, test_images / 255.0



# Definimos una lista con los nombres de las distintas clases que detecta la 
# red. Claro, la salida de la red sera un vector de numeros, donde  el numero 
# mas alto indicará cual es la clase que se ha detectado. Con esta lista, 
# podremos ponerle nombre a esa clase.
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']



# Mostramos algunas imagenes etiquetadas como ejemplo del dataset
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i])
    # The CIFAR labels happen to be arrays, 
    # which is why you need the extra index
    plt.xlabel(class_names[train_labels[i][0]])
plt.show()



# Creamos las capas de la red
model = tf.keras.models.Sequential([

    # Primero metemos las capas convolucionales.Este conjunto de capas se basa 
    # en dos tipos de capas, primero las capas convolucionales (Conv2D), las 
    # cuales se encargarán de hacer las convoluciones como hemos visto en la 
    # primera sesion; y las capas de pooling (MaxPooling2D), las cuales de 
    # momento solo sirven para reducir el tamaño de la imagen.
    #
    # No le deis mucha importancia a estas segundas, solamente hay que poner 
    # una de estas despues de una convolucion.
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),

    # Despues de las capas de convolucion nos quedará una imagen muy pequeña. 
    # Ahora solo hay que aplanarla para que entre en las capas de neuronas densas
    tf.keras.layers.Flatten(),

    # Por ultimo, las capas de neuronas densas, que son como las redes 
    # neuronales classic que hemos utilizado hasta ahora
    #tf.keras.Input(shape=x_train[0].shape),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10)
])



# Preparamos la funcion de loss
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)


# Compilamos el modelo
model.compile(optimizer='adam',
              loss=loss_fn,
              metrics=['accuracy'])


# Entrenamos
history = model.fit(train_images, train_labels, epochs=10, 
                    validation_data=(test_images, test_labels))



# Sacamos graficas de como ha ido el entrenamiento
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.show()


# Hacemos una evaluacion final de la red 
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)


# Sacamos algunos ejemplos de lo que predice la red, a ver que tal salen
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(test_images[i])

    # Sacamos la prediccion de la red para la imagen
    prep_img = np.array([test_images[i]])
    pred = model(prep_img)
    idx = np.argmax(pred)

    
    plt.xlabel(class_names[idx])
plt.show()
```
