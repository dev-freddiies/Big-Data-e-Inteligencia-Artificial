```python

### ··· Vamos a importar las librerias necesarias ··· ###

import speech_recognition as sr          # Libreria para el reconocimiento de la voz, que mas tarde hay que ponerla en language='es', porque si no es por defecto en inglés y no reconocerá nada que le mandemos
import pyttsx3                           # Libreria que nos permite escuchar lo que está escrito
import pywhatkit                         # Librería que nos servirá para enviar los whatsapp
import datetime                          # Libreria que nos sirve para saber la hora y fecha

import wikipedia                         # Para utilizar la wikipedia
wikipedia.set_lang("es")                 # Tenemos que cambiar el idioma a español (es) porque si no por defecto es en inglés

import pyjokes                           # Librería para que nos cuente una broma, mas tarde habrá que cambiarla a español, si no por defecto también es en inglés
import os                                # La libreria os permite realizar operaciones dependiente del Sistema Operativo como crear una carpeta, listar contenidos de una carpeta, conocer acerca de un proceso, finalizar un proceso, etc.
import smtplib                           # A traves del protocolo smtp podemos enviar emails, ademas de con la ayuda de esta libreria
from datetime import datetime

### ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ###
```

```python

### ··· Vamos a inicializar variables para la voz y trabajar mejor ··· ###

listener = sr.Recognizer()                    # Inicializamos la variable listener con la libreria para utilizarla mejor a la hora de que nos reconozca la voz
engine = pyttsx3.init()                       # Utilizamos la avriable engine, que va a ser nuestro motor de voz y la inicializamos para trabajar mejor con ella
voices = engine.getProperty ('voices')        
engine.setProperty('voice', voices[1].id)     # Seleccionamos la voz que queremos del array voices, en este caso el 1
engine.setProperty("rate", 150)               # Cambiamos la voz con el array, y le bajamos la velocidad para que se entienda mejor con el rate

### ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ###
```

```python

### ··· Funcion para que hable el programa ··· ###

def talk(text):            
    engine.say(text)
    engine.runAndWait()
    
### ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ###
```

```python

### ··· Funcion para que el asistente te de los buenos dias, tardes o noches, o basicamente que al arrancar el programa te diga lo que quieras ··· ###

def wishMe():

    hour = int(datetime.now().hour)
    
    if hour>=0 and hour<12:                                             # Depende de la hora que sea, nos dice una u otra cosa
        talk("¡Buenos días!")

    elif hour>=12 and hour<18:
        talk("¡Buenas tardes!")   

    else:
        talk("¡Buenas noches!")  

    talk("Hola, soy un asistente personal, me puedes llamar Jarvis")
    
### ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ###
```

```python

### ··· Funcion para que el asistente te pida lo que quieres hacer por voz ··· ###

#def take_command(command='hora'): --- Esto es para que arranque el programa directamente con algo para ser mas rapido a la hora de pronarlo

def take_command():
    
    try:
        with sr.Microphone() as source:
            print('Te estoy escuchando...')
            voice = listener.listen(source)
            command = listener.recognize_google(voice, language='es-es')     # Hay que cambiarlo a español porque si no no reconoce 
            command = command.lower()
            
    except Exception as e:   
        print("Repitelo, porfavor :) ...")  
        return "None"
    
    return command

### ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ###
```

```python

### ··· Funcion para que el asistente mande un correo ··· ###

def sendEmail(to, content):
    server = smtplib.SMTP('smtp.gmail.com', 587)                    # Enviamos correos por el protocolo smtp, o el puerto 587 que es el puerto del smtp
    server.ehlo()
    server.starttls()
    email = input('Introduce tu email: ')                           # Se ha configurado que se tenga que introducir el correo por escritura al igual que la contraseña para aumentar un poco la seguridad
    print(email)
    password = input('Introduce tu password: ')
    server.login(email, password)
    server.sendmail('alfredo98norba@gmail.com', to, content)        
    server.close()
    
### ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ###
```

```python

### ··· Vamos a iniciar el programa con los parametros que le demos ··· ###


bandera = True                                             # Utilizamos una bandera para evitar que se meta en un bucle infinito

if __name__ == "__main__":
    wishMe()                                               # Ejecutamos la funcion de los buenos dias o que nos diga lo que queramos
    
    while (bandera == True):
        
        command = take_command()                           # Esto es lo que reconoce que le decimos
        
        print(command)                                     # Nos muestra por pantalla la eleccion que le hemos dicho
        
        
        
        if 'reproduce' in command:                         # Si decimos por ejemplo: Reproduce Extremoduro abrirá YouTube con lo que queramos, en este caso nos abrirá una pestaña de YT con una canción de Extremoduro
            
            song = command.replace('reproduce', '')
            
            talk('Esta sonando: ' + song)
            print('Esta sonando: ' + song)
            
            pywhatkit.playonyt(song)                       # Utilizamos la libreria pywhatkit con la funcion playonyt de lo que dugamos despues de reproduce
            
            bandera = False                                # Salimos con la bandera
            
            
            
        elif 'hora' in command:                            # Para que nos diga la hora, decimos: hora
            
            time = datetime.now().strftime('%H:%M:%S')     # Hay que cambiar el formato de la hora a %H:%M:%S que es el español
            
            talk('La hora actual es: ' + time)             # Nos dice la hora hablada con talk
            print('La hora actual es: ' + time)
            
            bandera = False                                # Salimos con la bandera
            
            
            
        elif 'fecha' in command:                           # Para la fecha es igual que con la hora, la misma librería además, pero utilizamos una función diferente
            
            fecha = datetime.now().strftime('%d/%m/%Y')    # Hay que cambiar el formato de la fecha a dia, mes y año
             
            talk('La fecha actual es: ' + fecha)           # Nos lo dice de forma hablada
            print('La fecha actual es: ' + fecha)          # Y por pantalla
            
            bandera = False                                # Salimos con la bandera

            
       
        elif 'whatsapp' in command:                                        # Para enviar un whatsapp primero hay que meterse en otra pestaña y logearse en whatsapp
            
            try:
                
                talk("Por favor, di el numero")                            # Le decimos el numero por voz
                #content1 = '+34609208503' + take_command()                # También podemos inicializar directamente el numero 
                content1 = take_command()                                  # Con take command coge el numero
                
                talk("¿Que quieres decir'")                                # Aqui lo que queremos decir en el mensaje
                content5 = take_command()                                 
                pywhatkit.sendwhatmsg_instantly(content1, content5)        # Envia el mesaje al numero (content1) con lo que le hemos dicho (content5)
                
                talk("¡El mensaje se ha enviado!")
                
                bandera = False                                            # Salimos con la bandera

                
            except Exception as e:                                         # Si hay algun problema nos lo indica
                
                print(e)
                talk("Lo siento no se ha podido enviar el mensaje")
                
                bandera = False
                
                
                
        elif 'busca' in command:                         # Para buscar algo en wikipedia solo nos bastará decir busca y lo que queramos, por ejemplo, busca iphone, por ejemplo
                                                         # Porque el programa hace primero el busca y se va a la wikipedia y busca lo que le digamos posteriormente
            person = command.replace('busca', '')
            info = wikipedia.summary(person, 2)
            
            print(info)
            talk(info)                                   # Nos dice lo que hayamos querido buscar, que es un texto directamente sacado de la wikipedia
            
            bandera = False                              # Salimos con la bandera
                        


                
                
        elif 'cita' in command:                          # Podemos añadirle al programa lo que queramos, para que hable y nos diga lo que queramos e incluso que nos lleve a la pagina que le indiquemos
            talk('Lo siento no puedo, tengo futbol')
            
            bandera = False
            
        elif 'solteria' in command:
            talk('Estoy en una relacion')
            
            bandera = False
            
        elif 'gracias' in command:
            talk('De nada, estoy aqui para servir')
            
            bandera = False
            
        elif "quien te ha creado" in command or "quien es tu creador" in command: 
            talk("Me ha creado Alfredo Talavera Ramajo o... Tony Stark")
            
            bandera = False
            
        elif "discutir" in command: 
            talk("Vamos a discutir sobre mi creacion")
            
            bandera = False
            
            
            
            
        elif 'broma' in command:                                          # Para que nos cuente una broma solo tenemos que decir 'broma' y nos la dirá a parte que con el print nos la mostrará por pantalla, esto lo hace con la libreria pyjokes y la funcion get_jokes
            talk(pyjokes.get_joke(language="es", category="all"))         # Le cambiamos el idioma a español para que la broma sea en español y de cualquier categoria con all
            print(pyjokes.get_joke(language="es", category="all"))

            bandera = False
            
            

        elif 'email' in command:                                          # Para enviar un correo utiizaremos la anteriormente mencionada libreria de smtp y la funcion send_email
            try:
                talk("¿Que quieres poner en el mail?")                    # Le decimos lo que queremos decir en el correo
                content = take_command()
                
                to = "alfredo98norba@gmail.com"                           # Para que funcione un poco mejor le inicializamos el correo directamente, si no habria que cambiarlo e introducirlo hablando con talk
                
                sendEmail(to, content)                                    # Con la funcion sendEmail nos envia lo que le hemos dicho por voz a quien se lo hemos indicado
                
                talk("¡El mail se ha enviado satisfactoriamente!")        # En principio debería de ir bien y que nos indique que se ha enviado satisfactoriamente
                
                bandera = False
                
            except Exception as e:                                        # En cambio si hay algun error nos lo hará saber
                
                print(e)
                talk("Lo siento, no se ha podido enviar el mensaje") 
                
                bandera = False
                
                
                
        else:
            print("Perdona, no te he entendido")                          # Si no detecta ningun comando por voz nos dirá que no ha entendido nada y con la bandera a True se volverá a meter al bucle y nos volverá a pedir otro comandode voz como hora, whatsapp o etc
            
            bandera = True
            
### ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ··· ###
  
```
