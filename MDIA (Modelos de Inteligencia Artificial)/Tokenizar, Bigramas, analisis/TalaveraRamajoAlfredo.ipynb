```python
## Steamming, convertir cada palabra en su raiz
## para ello primero importamos las librerias y despues tokenizamos la frase
## una vez tokenizada la frase, utilizamos la función para obtener la raiz de las palabras.
## Utilizaremos el snowball, que es el mejor para textos en español
## se utiliza para clasificación de textos o en sistemas de recuperacion de la informacion.

from nltk.stem import SnowballStemmer
es_stem = SnowballStemmer('spanish')

stem_tokens = []

for tok in clean_tokens:
    stem_tokens.append(es_stem.stem(tok))

print(stem_tokens)
```

```python
# Generar dataframe que tenga esta estructura: PALABRA - RECUENTO - DENSIDAD
# hay q importar pandas
conteo = pd.DataFrame(columns=['PALABRA','RECUENTO','DENSIDAD'])

# Obtenemos la frecuencia de cada token
freq = nltk.FreqDist(clean_tokens) #Usamos la lista anterior de tokens sin stopwords

#recorremos el analisis de nltk y frecuencia
for key,val in freq.items():
    #key es el token
    #val es el numero de repeticiones del token
        
    #En el bucle insertamos las columnas
    dens="{:.2%}".format(val/len(clean_tokens))#obtenemos densidad
    reps="{:.0f}".format(val)#obtenemos repeticiones

    fila=pd.Series([key,reps, dens], index=conteo.columns)
    conteo=conteo.append(fila,ignore_index=True)

conteo.sort_values('DENSIDAD',ascending=False)
```

```python
#recibe contenido (frase) y clean booleano en el que indicamos
#si hacemos (true) o no (false) lematizacion
def text_density(content,clean):

#1 tokenizar
    tokenss= word_tokenize(content,"spanish")


    #2 eliminamos signos puntuacion y pasamos a minusculas
    tokens_alpha = [word.lower() for word in tokenss if  word.isalpha()]

    #Duplicamos los tokens
    clean_tokens = tokens_alpha[:]

    #3 eliminar stopwords
    for token in tokens_alpha:
        if token in stopwords.words('spanish'):
            clean_tokens.remove(token)
    # 4 Lematizacion: obtener lemas
    cleaned_tokens = []


    #si queremos lematizacion
    if clean:  
    #realizar la limpieza con SPACY
        separator= ' '
    #recorremos cada token
    for token in nlp(separator.join(clean_tokens)):
        cleaned_tokens.append(token.lemma_) #añadimos a la lista
    else:
        cleaned_tokens = clean_tokens #los tokens limpios son los que entraron


     # 5. Generar la tabla
    conteoo = pd.DataFrame(columns=['Palabra', 'Recuento','Densidad' ])
    
    #obtenemos la frecuencia de cada token
    freq =nltk.FreqDist(cleaned_tokens) #cleaned_tokens están tokenizados sin stopwords

    #recorremos el analisis de nltk y frecuencia
    for key,val in freq.items():
        #key es el token
        #val es el numero de repeticiones del token

        #En el bucle insertamos las columnas
        dens="{:.2%}".format(val/len(cleaned_tokens))#obtenemos densidad
        reps="{:.0f}".format(val)#obtenemos repeticiones

        fila=pd.Series([key,reps, dens], index=conteoo.columns)
        conteoo=conteoo.append(fila,ignore_index=True)

        result=conteoo.sort_values('Densidad',ascending=False)
    return result
```
